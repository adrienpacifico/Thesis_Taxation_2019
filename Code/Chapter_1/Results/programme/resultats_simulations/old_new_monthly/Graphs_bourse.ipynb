Les 10 plus gros volumes de transactions

```python
df_group_by = df.groupby(["label","date"])
biggest_volume = df.groupby(["label"])["volume"].sum()
```

```python
biggest_volume.sort(ascending=False); biggest_volume[:10]
```

```python
ten_biggest_volume_compagny = biggest_volume[:10].index.values ; ten_biggest_volume_compagny
```

```python
ten_biggest_volume_compagny = list(ten_biggest_volume_compagny)
```

```python
df.date = df.index
```

```python
df_group_by = df.groupby(["label","date"])
grouped_df = df_group_by.sum()
```

```python
grouped_df.loc[ten_biggest_volume_compagny]
```

```python
df_high = grouped_df.unstack(0).high.copy()
```

```python
df_high
```

```python
df_high[ten_biggest_volume_compagny].plot(figsize=(20,20)); plt.legend(loc='best')
```

```python
 grouped_df["average"] = (grouped_df.open + grouped_df.close)/2
```

```python
grouped_df.unstack(0).average["AXA"].plot()
```

Nomalized value

```python
first_day_average = grouped_df.query('date == "2001-08-06"').average
```

```python
first_day_average.index = first_day_average.index.droplevel(1); first_day_average
```

```python
grouped_df["normalised_value"] = grouped_df.average/first_day_average
```
